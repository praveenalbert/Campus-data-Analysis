#Preprocessing the data by filling the null values

```python
mean_value_1=data['inter_gpa'].mean()
mean_value_2=data['ssc_gpa'].mean()
mean_value_3=data['cgpa'].mean()
```

#correlation between the technical skills and the projects done by the students

###tier1

```python
data_pro = pd.DataFrame()
data_pro['no_of_projects']=data_tier1['no_of_projects']
data_pro['technical_skills'] = data_tier1['web_dev'] + data_tier1['Machine Learning'] + data_tier1['mobile_dev'] + data_tier1['cloud']
data_pro['technical_skills'] = data_pro['technical_skills'].astype(int)

# Calculate correlation coefficient
correlation_coefficient = data_pro['technical_skills'].corr(data_pro['no_of_projects'])

# Create a scatter plot
plt.scatter(data_pro['technical_skills'], data_pro['no_of_projects'], c='blue', alpha=0.7, label='Data Points')

# Add labels and title
plt.xlabel('Technical Skills')
plt.ylabel('Number of Projects')
plt.title(f'Correlation between Number of Projects and Technical Skills (Correlation: {correlation_coefficient:.2f})')

# Calculate the best-fit line (correlation line)
slope, intercept = np.polyfit(data_pro['technical_skills'], data_pro['no_of_projects'], 1)
best_fit_line = slope * data_pro['technical_skills'] + intercept

# Plot the best-fit line
plt.plot(data_pro['technical_skills'], best_fit_line, color='red', label='Correlation Line')
plt.xticks(np.unique(list(data_pro['technical_skills'])))
# Add a legend
plt.legend()

# Show the plot
plt.show()
```

From the above graphs we can observe that there is a positive correlation between technical skills and number of projects done by them <br><br>
If the number of Technical Skills known by the student is higher then the number of projects done by them would also be higher

CGPA Vs Salary


```python

cgpa = data_tier1['cgpa']
salary = data_tier1['salary_as_fresher']

# Calculate the average salary for each CGPA level
average_salary_by_cgpa = data_tier1.groupby('cgpa')['salary_as_fresher'].mean()
branch_colors = {
    'CSE': 'red',
    'ECE': 'green',
    'MECH': 'cyan',
    'EEE': 'purple',
    'IT': 'yellow'
}
branch_data={}
# Create a scatter plot
# plt.scatter(cgpa, salary, alpha=0.5, label='Individual Data')
plt.figure(figsize=(8, 5))
over_all_sal_tier1=list()
for branch, group in data_tier1.groupby('branch'):
    plt.scatter(group['cgpa'], group['salary_as_fresher'], alpha=0.5, label=branch, color=branch_colors[branch])
    over_all_sal_tier1=over_all_sal_tier1+ group['cgpa'].tolist()
    branch_data[f"{branch}_salary"] = group['salary_as_fresher'].tolist()
over_all_sal_tier1.sort()

# Plot the average line
plt.axhline(average_salary_by_cgpa.mean(), color='red', linestyle='dashed', linewidth=2, label='Average Salary')
branch_wise_list={}
for branch, branch_dict in branch_data.items():
  branch_wise_list[branch]=branch_dict
# Add labels and title
plt.xlabel('CGPA')
plt.ylabel('Salary')
plt.title('CGPA vs. Salary Tier1')
plt.legend()

# Show the plot
plt.show()
```
