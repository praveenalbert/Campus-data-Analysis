## EDA

```python
import random
n=11
cs = ["#" + ''.join([random.choice('0123456789ABCDEF')
                        for j in range(6)]) for i in range(n)]
```

```python
plt.pie(placed_corr,labels=placed_corr.index,autopct='%.2f',colors=cs)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.tight_layout()
plt.title("percentage of Factors Affecting placement")
plt.show()
```

```python
#using kendall
placed_corr_1=data.corr(method="kendall").Placed[:-1]
placed_corr_1
```

```python
plt.pie(placed_corr_1,labels=placed_corr_1.index,autopct='%.2f',colors=cs)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.tight_layout()
plt.title("percentage of Factors Affecting placement")
plt.show()
```
